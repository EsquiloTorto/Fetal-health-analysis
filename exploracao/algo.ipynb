```python
import pandas as pd
import numpy as np
```

```python
fetal = pd.read_csv('../dataset/fetal_health.csv')
```

```python
fetal.head(10)
```

```python
fetal.corr()
```

```python
# do a correlation heatmap
import seaborn as sns
import matplotlib.pyplot as plt

data = fetal[['accelerations','prolongued_decelerations', 'abnormal_short_term_variability', 'histogram_mean', 'histogram_variance', 'fetal_health']]

plt.figure(figsize=(15,15))

sns.heatmap(data.corr(), annot=True, cmap='coolwarm')

plt.show()
```

```python
# make a splom plot
from pandas.plotting import scatter_matrix

scatter_matrix(fetal, figsize=(15,15))

# show y labels horizontally
plt.yticks(rotation=0)

# show x labels vertically
plt.xticks(rotation=90)

plt.show()
```

```python
# get all columns except the ones in [11:20]
colunas = list(fetal.iloc[:, 11:21].columns)

# from fetal data select all columns except the ones in colunas
data2 = fetal.drop(columns=colunas)

scatter_matrix(data2, figsize=(15,15))

plt.show()
```

```python
sns.heatmap(data2.corr(), annot=True, cmap='coolwarm')

plt.show()
```

```python
# scatterplot of abnormal_short_term_variability and mean_value_of_long_term_variability
plt.scatter(fetal['abnormal_short_term_variability'], fetal['mean_value_of_long_term_variability'])

plt.show()
```
